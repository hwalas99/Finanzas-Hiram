# Finanzas – Semana 1: Interés compuesto


1. Si inviertes $10,000 MXN a una tasa del 12% anual capitalizada anualmente durante 5 años, ¿cuánto tendrás al final del periodo?

2. ¿Cuánto necesitas aportar mensualmente durante 3 años (36 meses) para alcanzar una meta de $100,000 MXN, si tu dinero crece al 8% anual compuesto mensualmente?

Utiliza fórmulas financieras en Python (sin librerías externas). Redondea a 2 decimales.


```

# Ejercicio 1: Interés compuesto anual
P = 10000  # capital inicial
r = 0.12   # tasa anual
t = 5      # años

VF = P * (1 + r)**t
print(f"Valor futuro al final de 5 años: ${VF:.2f}")
```

```

# Ejercicio 2: Aportaciones mensuales para alcanzar una meta
meta = 100000
r_anual = 0.08
r_mensual = r_anual / 12
n = 36  # meses

# Fórmula de anualidad ordinaria: A = FV * r / ((1 + r)^n - 1)
A = meta * r_mensual / ((1 + r_mensual)**n - 1)
print(f"Aportación mensual necesaria: ${A:.2f}")
```
